```python
import numpy as np

products = []

for even in range(2, 21, 2):
    odd = even - 1
    product = even * odd
    products.append(product)

products_array = np.array(products)

print(products_array)
```

```python
import numpy as np

random_integers = np.random.randint(1, 101, size=(3, 3))

squared_matrix = np.square(random_integers)

print("Матриця з випадкових чисел:")
print(random_integers)
print("\nМатриця квадратів цих чисел:")
print(squared_matrix)
```

```python
import numpy as np

random_floats = np.random.uniform(-2, 2, size=(5, 5))

mean_absolute_value = np.mean(np.abs(random_floats))

print("Масив випадкових дійсних чисел від -2 до 2:")
print(random_floats)
print("\nСереднє за модулем значення елементів масиву:")
print(mean_absolute_value)
```

```python
import numpy as np

random_floats = np.random.uniform(0, 1, size=(5, 5))

mean_value = np.mean(random_floats)

elements_less_than_mean = random_floats[random_floats < mean_value]

sum_of_elements = np.sum(elements_less_than_mean)

print("Масив випадкових дійсних чисел від 0 до 1:")
print(random_floats)
print("\nСереднє значення масиву:")
print(mean_value)
print("\nЕлементи, менші за середнє значення:")
print(elements_less_than_mean)
print("\nСума елементів, менших за середнє значення:")
print(sum_of_elements)
```

```python
import numpy as np

array1 = np.random.uniform(-2, 2, size=7)
array2 = np.random.uniform(-2, 2, size=7)

addition = array1 + array2

subtraction = array1 - array2

multiplication = array1 * array2

print("Перший масив:")
print(array1)
print("\nДругий масив:")
print(array2)
print("\nРезультат поелементного додавання:")
print(addition)
print("\nРезультат поелементного віднімання:")
print(subtraction)
print("\nРезультат поелементного множення:")
print(multiplication)
```

```python
import numpy as np

matrix1 = np.random.randint(7, 18, size=(4, 4))
matrix2 = np.random.randint(7, 18, size=(4, 4))

matrix_product = np.dot(matrix1, matrix2)

print("Перша матриця:")
print(matrix1)
print("\nДруга матриця:")
print(matrix2)
print("\nДобуток матриць:")
print(matrix_product)
```

```python
import numpy as np

matrix = np.random.randint(1, 11, size=(2, 2))

inverse_matrix = np.linalg.inv(matrix)

print("Матриця 2x2:")
print(matrix)
print("\nОбернена матриця:")
print(inverse_matrix)
```

```python
import numpy as np

matrix = np.random.uniform(10, 50, size=(3, 3))

transposed_matrix = np.transpose(matrix)

print("Матриця 3x3:")
print(matrix)
print("\nТранспонована матриця:")
print(transposed_matrix)
```

```python
import numpy as np

matrix = np.random.randint(1, 11, size=(3, 4))

vector = np.random.uniform(0, 1, size=(4, 1))

matrix_vector_product = np.dot(matrix, vector)

print("Матриця 3x4:")
print(matrix)
print("\nВектор 4x1:")
print(vector)
print("\nДобуток матриці на вектор:")
print(matrix_vector_product)
```

```python
import numpy as np

matrix1 = np.random.randint(1, 6, size=(3, 3))

matrix2 = np.random.randint(1, 6, size=(3, 3))

diagonal_product1 = np.prod(np.diag(matrix1))

diagonal_product2 = np.prod(np.diag(matrix2))

print("Перша матриця 3x3:")
print(matrix1)
print("\nДобуток елементів діагоналі першої матриці:", diagonal_product1)

print("\nДруга матриця 3x3:")
print(matrix2)
print("\nДобуток елементів діагоналі другої матриці:", diagonal_product2)
```

```python
import numpy as np

matrix1 = np.random.uniform(-1, 1, size=(3, 3))

matrix2 = np.random.uniform(-1, 1, size=(3, 3))

sum_matrix1 = np.sum(matrix1)

sum_matrix2 = np.sum(matrix2)

print("Перша матриця 3x3:")
print(matrix1)
print("\nСума елементів першої матриці:", sum_matrix1)

print("\nДруга матриця 3x3:")
print(matrix2)
print("\nСума елементів другої матриці:", sum_matrix2)
```

```python
import numpy as np

matrix = np.random.uniform(-3, 3, size=(3, 3))

max_element = np.max(matrix)

print("Матриця 3x3:")
print(matrix)
print("\nНайбільший елемент матриці:", max_element)
```

```python
import numpy as np

matrix = np.random.uniform(-0.5, 0.5, size=(3, 3))

zero_counts = np.sum(matrix == 0, axis=1)

print("Матриця 3x3:")
print(matrix)
print("\nКількість нульових елементів в кожному рядку:")
print(zero_counts)
```
